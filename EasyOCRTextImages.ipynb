```python
import easyocr # To perform Optical Character Recognition
import cv2 # OpenCV = To show & visualize image
from matplotlib import pyplot as plt
import numpy as np
```

# Read in images or video

```python
IMAGE_PATH = '/content/sample_data/text_logo_11.png'
# Using easyocr to perform Optical Character Recognition
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(IMAGE_PATH)
result
# This entire array is defining where the text is in our image, then the text which is identified in the image i.e. FedEx
# Then is the confidence
```

# Draw Results

```python
# Here we're giving the input as tuple cuz OPENCV always expect a tuple argument
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(IMAGE_PATH) # reading the image path
img = cv2.rectangle(img, top_left, bottom_right, (0,0,255), 5) # no. of additional visualizations like bounded rectangle
img = cv2.putText(img, text, top_left, font, .5, (255,0,255), 2, cv2.LINE_AA) # For detecting the text and writing on it in pink color
plt.figure(figsize=(15,15)) # Enlarging our image
plt.imshow(img) # To show the transformed image
plt.show() # To print
```

# Trying with other images

```python
IMAGE_PATH = '/content/sample_data/SSW-OOS-1.jpg'
# Using easyocr to perform Optical Character Recognition
reader = easyocr.Reader(['en'], gpu=False)
result = reader.readtext(IMAGE_PATH)
result
# This entire array is defining where the text is in our image, then the text which is identified in the image i.e. FedEx
# Then is the confidence
```

```python
# Here we're giving the input as tuple cuz OPENCV always expect a tuple argument
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
# For handling multiple line of texts
img = cv2.imread(IMAGE_PATH)
for detection in result:
  top_left = tuple([int(val) for val in detection[0][0]])
  bottom_right = tuple([int(val) for val in detection[0][2]])
  text = detection[1]
  font = cv2.FONT_HERSHEY_SIMPLEX
  img = cv2.rectangle(img, top_left, bottom_right, (0,0,255), 4) # no. of additional visualizations like bounded rectangle
  img = cv2.putText(img, text, top_left, font, 2, (255,0,255), 1, cv2.LINE_AA) # For detecting the text and writing on it in pink color
plt.figure(figsize=(15,15)) # Enlarging our image
plt.imshow(img)
plt.show()
```
